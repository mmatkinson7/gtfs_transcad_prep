```python
import numpy as np
import pandas as pd
import geopandas as gpd
# import shapes shapefile (post update from arcpro)
```

```python
fred = gpd.read_file(r"C:\Users\matkinson.AD\Downloads\shapes_edited_shp.shp")
fred
```

```python
columns = ["shape_id","shape_pt_lat","shape_pt_lon","shape_pt_sequence","shape_dist_traveled"]
fred = fred[['shape_id','geometry']]
fred
```

```python
fred['points'] = fred.apply(lambda x: [y for y in x['geometry'].coords], axis = 1)
```

```python
fred
```

```python
def coords_lines(df, sid):
    x,y = df.iloc[df.index[fred['shape_id'] == sid].to_list()[0]].geometry.coords.xy
    z = pd.DataFrame({'shape_id': sid,'shape_pt_lat':y,'shape_pt_lon':x})
    z['shape_pt_sequence'] = z.index + 1
    return z

coords_lines(fred, "010038")
```

```python
#fred.apply(lambda x: coords_lines(x['shape_id']),axis=1)
count = 0
for x in fred['shape_id']:
    if count == 0:
        z = coords_lines(fred, x)
    else:
        z = pd.concat([z,(coords_lines(fred,x))])
    count += 1
z = z.reset_index()[['shape_id', 'shape_pt_lat','shape_pt_lon', 'shape_pt_sequence']]
z
```
